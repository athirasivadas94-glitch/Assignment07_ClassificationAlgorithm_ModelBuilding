```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
from sklearn.preprocessing import StandardScaler


# Load dataset
data = load_breast_cancer()

# Convert to DataFrame
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
```

```python
# Check null values
df.isnull().sum()
```

After checking the dataset using isnull().sum(), we found that the breast cancer dataset does not contain any missing values.
This step is important because missing values can negatively affect model performance and lead to incorrect predictions if not handled properly.

```python
# Separate features and target
X = df.drop('target', axis=1)
y = df['target']

# Apply Standard Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

We applied StandardScaler to scale the features so that they have a mean of 0 and a standard deviation of 1.
This is necessary because:

The dataset contains features with different ranges (e.g., radius, area, texture).

Many machine learning algorithms like Logistic Regression, SVM, and KNN perform better when features are on the same scale.

Scaling improves model accuracy and convergence speed.

The breast cancer dataset contains multiple numerical features with varying magnitudes. Without scaling, features with larger values (like area or perimeter) could dominate the learning process. Proper preprocessing ensures fair contribution of all features and improves the overall performance of supervised learning models.

**Classification Algorithm Implementation**
Implement the following five classification algorithms:
1. Logistic Regression
2. Decision Tree Classifier
3. Random Forest Classifier
4. Support Vector Machine (SVM)
5. k-Nearest Neighbors (k-NN)
For each algorithm, provide a brief description of how it works and why it might be suitable for this dataset.

```python
#Split Dataset
from sklearn.model_selection import train_test_split

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

**1. Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression

# Initialize and train model
log_reg = LogisticRegression(max_iter=5000)
log_reg.fit(X_train, y_train)

# Predictions
y_pred_lr = log_reg.predict(X_test)
```

Logistic Regression is a linear classification algorithm that predicts the probability of a class using a sigmoid function and then classifies the output into binary categories.

The breast cancer dataset is a binary classification problem,Logistic Regression works well for linearly separable data and provides good interpretability, making it suitable for medical datasets.

**2. Decision Tree Classifier**

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize and train model
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)

# Predictions
y_pred_dt = dt.predict(X_test)
```

Decision Tree Classifier splits the data into branches based on feature values, creating a tree-like structure to make decisions and classify data points.

It can handle complex relationships between features such as radius, texture, and area. It also does not require feature scaling and is easy to interpret, which is useful for healthcare-related analysis.

**3. Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize and train model
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Predictions
y_pred_rf = rf.predict(X_test)
```

Random Forest is an ensemble learning method that builds multiple decision trees and combines their predictions to improve accuracy and reduce overfitting.

The dataset contains many features, and Random Forest can handle high-dimensional data effectively. It improves prediction accuracy and reduces overfitting compared to a single decision tree.

**4. Support Vector Machine (SVM)**

```python
from sklearn.svm import SVC

# Initialize and train model
svm = SVC(kernel='linear')
svm.fit(X_train, y_train)

# Predictions
y_pred_svm = svm.predict(X_test)
```

Support Vector Machine finds the optimal hyperplane that separates different classes by maximizing the margin between them.

SVM performs well on high-dimensional datasets like the breast cancer dataset. Since the data is scaled and mostly linearly separable, SVM can achieve high classification accuracy.

**5. k-Nearest Neighbors (k-NN)**

```python
from sklearn.neighbors import KNeighborsClassifier

# Initialize and train model
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Predictions
y_pred_knn = knn.predict(X_test)
```

k-NN is a non-parametric algorithm that classifies a data point based on the majority class of its k nearest neighbors using distance metrics.

Since the dataset is well-structured and scaled, k-NN can effectively measure distances between data points. It works well for classification tasks where similar feature patterns indicate similar classes, such as tumor characteristics in this dataset.

**Model Comparison **
● Compare the performance of the five classification algorithms.
● Which algorithm performed the best and which one performed the worst?

```python
from sklearn.metrics import accuracy_score

# Calculate accuracy for each model
accuracy_lr = accuracy_score(y_test, y_pred_lr)
accuracy_dt = accuracy_score(y_test, y_pred_dt)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
accuracy_svm = accuracy_score(y_test, y_pred_svm)
accuracy_knn = accuracy_score(y_test, y_pred_knn)

print("Logistic Regression Accuracy:", accuracy_lr)
print("Decision Tree Accuracy:", accuracy_dt)
print("Random Forest Accuracy:", accuracy_rf)
print("SVM Accuracy:", accuracy_svm)
print("k-NN Accuracy:", accuracy_knn)
```

Best Performing Algorithm: Logistic Regression
Worst Performing Algorithm :Decision Tree and k-NN

All models performed well with accuracy above 94%, which shows that the breast cancer dataset is well-structured and suitable for classification tasks. However, Logistic Regression achieved the highest accuracy, making it the most effective model for this dataset, while Decision Tree and k-NN showed comparatively lower performance.
